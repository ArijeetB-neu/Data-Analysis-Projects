## (Visualization 1)

> What is the distribution of the trip duration?

```python
bin = 10**np.arange(0 , 3+0.1 , 0.1)
ticks =  [ 0.1 , 0.3 , 1 , 3, 10, 30, 100, 300]
labels = ['{}'.format(v) for v in ticks]
plt.hist(data = df , x = 'duration_min' , bins = bin);
plt.xticks(ticks,labels);
plt.xlabel('Duration_min');
plt.xscale('log');
plt.title('Distribution of Trip Duration(minutes)')
plt.ylabel('Frequency');
```

> Conclusion 1: From the plot, we can say that most people tend to use the bikes for short-time trips ( around 10 mins).

## (Visualization 2)

> Question 2 : Whats the distrubition of bike rides as per user age group?

```python
#Generate a new field for member age group from member_age_bin
df['member_age_bin'] = df['member_age'].apply(lambda x: '10 - 20' if 10<x<=20
                                                  else '20 - 30' if 20<x<=30
                                                  else '30 - 40' if 30<x<=40
                                                  else '40 - 50' if 40<x<=50
                                                  else '50 - 60' if 50<x<=60
                                                  else x)
```

```python
df_trip_by_age = df.groupby('member_age_bin').agg({'bike_id':'count'})
```

```python
df_trip_by_age['perc'] = (df_trip_by_age['bike_id']/df_trip_by_age['bike_id'].sum())*100
```

```python
user_color = ['blue', 'green', 'green', 'red', 'red']
df_trip_by_age['perc'].plot(kind='bar', color=user_color, figsize=(12,8))
plt.title('Percentage of bike rides as per age group', fontsize=22, y=1.015)
plt.xlabel('member age group', labelpad=16)
plt.ylabel('pecentage(%) [rides]', labelpad=16)
plt.xticks(rotation=360);
```

> Conclusion 2 : 20 to 40 years old people took more than %70 of bike rides. Among those, 30 to 40 years old people's rides account almost %40 of all bike rides.

## (Visualization 3)
> Question 3 : Whats the distrubition of bike rides as per gender?

```python
df_trip_by_gender = df.groupby('member_gender').agg({'bike_id':'count'})
```

```python
df_trip_by_gender['perc'] = (df_trip_by_gender['bike_id']/df_trip_by_gender['bike_id'].sum())*100
```

```python
user_color = ['blue', 'green', 'black']
df_trip_by_gender['perc'].plot(kind='barh', color=user_color, figsize=(12,8))
plt.title('Percentage of bike rides as per age gender', fontsize=22, y=1.015)
plt.ylabel('member gender', labelpad=16)
plt.xlabel('pecentage(%) [rides]', labelpad=16)
plt.xticks(rotation=360)
plt.xlim(0,100);
```

> Conclusion 3 : Male took more than %70 of all bike rides, while female took around %24 of them.

## (Visualization 4)

Question 4. Does month affect the trip duration ?

```python
# Check the trip duration change over each month.
plt.figure(figsize = [10,8]) 
plt.scatter(data = df , x = 'start_time_month' , y = 'duration_min' , alpha = 0.1 )
plt.title('The Trip Duration Trend over each Month ')
plt.xlabel('Month')
plt.ylabel('Trip Duration(minutes)');
```

Question 5. Will season show a more obvious affect on trip duration ?

```python
# Create the 4 season categories
spring_time = df.query('start_time_month in ["3" , "4" , "5"]' ).duration_min.mean ()
summer_time = df.query('start_time_month in ["6" , "7" , "8"] ').duration_min.mean()
fall_time = df.query('start_time_month in ["9" , "10" , "11"] ').duration_min.mean()
winter_time = df.query('start_time_month in ["12" , "1" , "2"]' ).duration_min.mean()
# month must be double quoteed characters to avoid syntax errors
plt.bar(['spring','summer','fall','winter'],[spring_time , summer_time , fall_time , winter_time] )
plt.title('Trip Duration vs Seasons ')
plt.xlabel('Seasons')
plt.ylabel('Trip Duration');
```

>Conclusion 4 : Monthwise June to August seems to have the highest trip duration. Also, weather has somewhat limited affect over the trip duration. Summer and Spring have relatively higher trip duration. Winter turns to have the lowest trip duration. But this dip is very small owing to the temperate climate of SF Bay area.

## (Visualization 5)

Question 6 : Is trip duration affected by user type?  

```python
g = sb.FacetGrid(data =  df, hue = 'user_type', size = 6)
g.map(plt.scatter, 'start_time_month','duration_min', alpha = 1/3)
g.add_legend()
plt.title('Trip Duration(min) Against Month and User Type')
plt.xlabel('Month')
plt.ylabel('Trip Duration(min)');
```

> Conclusion 6: Its pretty much evident that there are far more customers than subscribers. Although the number of subscribers are quite less but they tend to rent the bikes for long trips, almost everytime above 250 minutes.
